```python
from sklearn.model_selection import train_test_split
x=df.drop('Class',axis=1)
y=df.Class
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
pred=dt.predict(xtest)
from sklearn.metrics import*
print('MAE ',mean_absolute_error(pred,ytest))
print('MSE ',mean_squared_error(pred,ytest))
print('RMSE ',np.sqrt(mean_squared_error(pred,ytest)))
```

```python
print(accuracy_score(pred,ytest))
print(confusion_matrix(pred,ytest))
print(classification_report(pred,ytest,digits=5))
```
