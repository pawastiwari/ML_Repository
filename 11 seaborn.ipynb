```python
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
sns.distplot(tips['total_bill'],bins=30)
```


# MATRIX PLOT

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc=tips.corr()
print(tc)
sns.heatmap(tc,annot=True)
```

```python
f=flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(f)
```

```python
sns.clustermap(f)
```

# GRID

```python
import seaborn as sns
%matplotlib inline 
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g=sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')  #VERY USEFUL
#g.map(sns.distplot,'total_bill')
g.map(sns.scatterplot,'total_bill','tip')
```

# REGRESSION

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
           scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='sex'
          ,aspect=0.4,size=10)
```

# STYLE AND COLOR

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
#these syntax applies globally needs to change if required
#sns.set_style('white')
#sns.set_style('ticks')
sns.set_style('darkgrid')
#sns.set_style('whitegrid')
sns.despine(right=False)

sns.countplot(x='sex',data=tips)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=1.5)
sns.countplot(x='sex',data=tips)
```

```python
#we have to change as compare to previous cell as this syntax applies globally
sns.set_context('notebook',font_scale=1.5)

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
